## Calculating the Pearson Coeficient
___


```python
from  scipy.stats.stats import pearsonr
import seaborn as sns
```

### Pearson Correlation

```python
X = cars[['mpg','hp','wt','qsec']]
sns.pairplot(X)
```

### Using scipy to calculate the Pearson Correlation Coeficient

```python
mpg = cars['mpg']
hp = cars['hp']
qsec = cars['qsec']
wt = cars['wt']

pearson_coefficient, p_value = pearsonr(mpg, hp)
print('Pearson Coefficient {:0.3}'.format(pearson_coefficient))
pearson_coefficient, p_value = pearsonr(mpg, qsec)
print('Pearson Coefficient {:0.3}'.format(pearson_coefficient))
pearson_coefficient, p_value = pearsonr(mpg, wt)
print('Pearson Coefficient {:0.3}'.format(pearson_coefficient))
```

### Using Pandas to calculate the Pearson Correlation Coeficient

```python
corr = X.corr()
corr
```

### Visualizing the Pearson Coefficient with Seaborn

```python
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values)
```

## Spearman Correlation
___


```python
from scipy.stats import spearmanr
```

```python
X = cars[['cyl','vs','am','gear']]
sns.pairplot(X)
```

```python
cyl = X['cyl']
vs = X['vs']
am = X['am']
gear = X['gear']

spearman_coef, p_value = spearmanr(cyl,vs)
print('Spearman Coefficient: {:0.3} p-value: {:0.3}'.format(spearman_coef,p_value))

spearman_coef, p_value = spearmanr(cyl,am)
print('Spearman Coefficient: {:0.3} p-value: {:0.3}'.format(spearman_coef,p_value))

spearman_coef, p_value = spearmanr(cyl,gear)
print('Spearman Coefficient: {:0.3} p-value: {:0.3}'.format(spearman_coef,p_value))
```

### Squared-Chi

```python
from scipy.stats import chi2_contingency
```

```python
tab = pd.crosstab(cyl,am)
chi2, p, dof, exp = chi2_contingency(tab.values)
print('Squared-chi: {:0.3} p-value: {:0.3}'.format(chi2,p))

tab = pd.crosstab(cyl,vs)
chi2, p, dof, exp = chi2_contingency(tab.values)
print('Squared-chi: {:0.3} p-value: {:0.3}'.format(chi2,p))

tab = pd.crosstab(cyl,gear)
chi2, p, dof, exp = chi2_contingency(tab.values)
print('Squared-chi: {:0.3} p-value: {:0.3}'.format(chi2,p))
```

## Pre-processing with Scikit-Learn
___


```python
from sklearn import preprocessing
```

### Normalizing and transforming variables with MinMaxScaler() and fit_transform()

```python
cars[['mpg']].describe()
```

```python
import matplotlib.pyplot as plt
mpg = cars.mpg.values
plt.plot(mpg)
```

```python
mpg_mat = mpg.reshape(-1,1)
scaler = preprocessing.MinMaxScaler()
scaled_mpg = scaler.fit_transform(mpg_mat)
plt.plot(scaled_mpg)
```

```python
scaler = preprocessing.MinMaxScaler(feature_range=(0,10))
scaled_mpg = scaler.fit_transform(mpg_mat)
plt.plot(scaled_mpg)
```

### Using scale() to standardize variables

```python
mpg_std = preprocessing.scale(mpg,axis=0,with_mean=False,with_std=False)
plt.plot(mpg_std)
```

```python
mpg_std = preprocessing.scale(mpg)
plt.plot(mpg_std)
```
