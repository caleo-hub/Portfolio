### Importing libraries and Classes

```python
import pandas as pd
```

```python
path = '../data/mtcars.csv'
cars = pd.read_csv(path)
cars.head()
```

### Renaming columns

```python
cars.columns = ['names', 'mpg', 'cyl', 'disp','hpt','drat',
               'wt','qsec','vs','am','gear','carb']
cars.head()
```
