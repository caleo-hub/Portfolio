## Concatenating Data
___

### Importing libraries and Classes

```python
import pandas as pd
import numpy as np

from pandas import DataFrame
```

```python
df = DataFrame(np.arange(36).reshape(6,6))
df
```

```python
df2 = DataFrame(np.arange(15).reshape((5,3)))
df2
```

### Concatenating Data

pd.concat([left_object,right_object])

```python
pd.concat([df, df2])
```

pd.concat([left_object,right_object], axis=1)

```python
pd.concat([df,df2], axis=1)
```
