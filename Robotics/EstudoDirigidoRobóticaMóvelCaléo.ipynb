## Modelo Cinemático I

$$ \begin{bmatrix}
X' \\
Y' \\
\theta'
\end{bmatrix} = 

\begin{bmatrix}
cos(\theta) & -sin(\theta) & 0\\
sin(\theta) & cos(\theta) & 0\\
0           & 0           & 1
\end{bmatrix}  

\begin{bmatrix}
1 & 0\\
0 & 0\\
0 & 1
\end{bmatrix}  

\begin{bmatrix}
r_d/2 & r_e/2\\
r_d/2l & -r_e/2l
\end{bmatrix}  

\begin{bmatrix}
\omega_d\\
\omega_r
\end{bmatrix}  
$$

E através da integração:

$$ \begin{bmatrix}
X(k+1) \\
Y(k+1) \\
\theta(k+1)
\end{bmatrix} = 


\begin{bmatrix}
X(k) \\
Y(k) \\
\theta(k)
\end{bmatrix} 

+

 \begin{bmatrix}
X' \\
Y' \\
\theta'
\end{bmatrix} * \delta t
$$

___

## Para velocidades angulares Wd = We = 1

```python
velocidadeAngularRodaDireita = 1
velocidadeAngularRodaEsquerda = 1

#Simulação no Coppelia Robotics
PosicaoRealHistorico = Simula(velocidadeAngularRodaDireita,
                                velocidadeAngularRodaEsquerda,
                                Passo=0.05,
                                TempoSimulacaoSegundos=TempoSimulacaoSegundos)


# Calculo do Modelo I
roboUm = RoboPionnerModeloUm()
PosicaoRoboHistorico = roboUm.posicaoRobo


for i in range(0,NumeroIteracoes-1):
    roboUm.velocidadeAngularRodas = [velocidadeAngularRodaDireita,velocidadeAngularRodaEsquerda]
    PosicaoNova = roboUm.calculaNovaPosicaoRobo(Passo=0.05)
    PosicaoRoboHistorico = np.append(PosicaoRoboHistorico,PosicaoNova,axis=1)

# Desenhando Gráficos
plt.rcParams["figure.figsize"] = (10,10)
plotaGrafico(PosicaoRoboHistorico,PosicaoRealHistorico)

```

## Modelo Cinemático II


![Dimensões Robô Modelo II](DimensoesRoboModeloII.png)

$$
R = {L \over 2} * {v_e + v_d \over v_e - v_d } \\

\omega =  {v_e - v_d \over L} \\

ICC = \begin{bmatrix} 
ICC_x \\
ICC_y \\

\end{bmatrix} 

= 
\begin{bmatrix} 
X - Rsin(\theta) \\
Y + Rcos(\theta) 

\end{bmatrix} \\


\begin{bmatrix}
X(k+1) \\
Y(k+1) \\
\theta(k+1)
\end{bmatrix} = 


\begin{bmatrix}
cos(\omega \delta t) & -sin(\omega \delta t) & 0\\
sin(\omega \delta t) & cos(\omega \delta t) & 0\\
0           & 0           & 1
\end{bmatrix}  


\begin{bmatrix}
X(k) - ICC_x \\
Y(k) - ICC_y \\
\theta(k)
\end{bmatrix} 

+

 \begin{bmatrix}
ICC_x \\
ICC_y \\
\omega \delta t
\end{bmatrix} 
$$

# Considerações

O Modelo Cinemático I do estudo dirigido é um modelo simples que ajuda dar uma orientação de como as dimensões do robô irão interferir em seu movimento, podendo ser implementado e utilizado rapidamente para uma modelagem inicial. 

Em contrapartida, o modelo Cinemático II baseado em odometria pode ser de utilizado como um modelo mais preciso para ao robô Pioneer. Mesmo utilizando somente as equações pré-estabelecidas, o modelo II possui uma leve vantagem na precisão de sua trajetória prevista.

Ambos os modelos nesse estudos possuem erros causados pelas variabilidades de um robô na realidade (nesse caso, o simulador). Questões desprezadas nas equações como derrapagem das rodas, atrito e variabilidade nas velocidades angulares dos motores prejudicam a eficiência dos modelos.

Contudo, o modelo cinemático II possui sua vantagem pela possibilidade ter o encoder óptico incorporado na sua modelagem, deixando mais preciso o cálculo da Velocidade Angular do robô e potencializando a precisão dos modelos.
